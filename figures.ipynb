```python
# distance cc and fi
p5cc_d = Rp_mat['Rp_cc_p5_d']
p20cc_d = Rp_mat['Rp_cc_p20_d']
p35cc_d = Rp_mat['Rp_cc_p35_d']
p5fi_d = Rp_mat['Rp_fi_p5_d']
p20fi_d = Rp_mat['Rp_fi_p20_d']
p35fi_d = Rp_mat['Rp_fi_p35_d']

# Student-Gumbel cc40 and fi55
p5_tg40 = Rp_mat['Rp_cc_p5_tg40']
p20_tg40 = Rp_mat['Rp_cc_p20_tg40']
p35_tg40 = Rp_mat['Rp_cc_p35_tg40']
p5_tg55 = Rp_mat['Rp_fi_p5_tg55']
p20_tg55 = Rp_mat['Rp_fi_p20_tg55']
p35_tg55 = Rp_mat['Rp_fi_p35_tg55']

# Gumbel cc40 and fi55
p5_gu40 = Rp_mat['Rp_cc_p5_gu40']
p20_gu40 = Rp_mat['Rp_cc_p20_gu40']
p35_gu40 = Rp_mat['Rp_cc_p35_gu40']
p5_gu55 = Rp_mat['Rp_fi_p5_gu55']
p20_gu55 = Rp_mat['Rp_fi_p20_gu55']
p35_gu55 = Rp_mat['Rp_fi_p35_gu55']

# Student cc40 and fi55
p5_st40 = Rp_mat['Rp_cc_p5_st40']
p20_st40 = Rp_mat['Rp_cc_p20_st40']
p35_st40 = Rp_mat['Rp_cc_p35_st40']
p5_st55 = Rp_mat['Rp_fi_p5_st55']
p20_st55 = Rp_mat['Rp_fi_p20_st55']
p35_st55 = Rp_mat['Rp_fi_p35_st55']

# SP500
RmRf = Rp_mat['RmRf']
```

```python
# cumulative excess returns

# distance cc and fi 
ret_p5cc_d = (1 + p5cc_d).cumprod() 
ret_p20cc_d = (1 + p20cc_d).cumprod()  
ret_p35cc_d = (1 + p35cc_d).cumprod()  
ret_p5fi_d = (1 + p5fi_d).cumprod()  
ret_p20fi_d = (1 + p20fi_d).cumprod()  
ret_p35fi_d = (1 + p35fi_d).cumprod()  

# Student-Gumbel cc40 and fi55
ret_p5cc_tg40 = (1 + p5_tg40).cumprod()  
ret_p20cc_tg40 = (1 + p20_tg40).cumprod()  
ret_p35cc_tg40 = (1 + p35_tg40).cumprod()  
ret_p5fi_tg55 = (1 + p5_tg55).cumprod()  
ret_p20fi_tg55 = (1 + p20_tg55).cumprod()  
ret_p35fi_tg55 = (1 + p35_tg55).cumprod()  

# Gumbel cc40 and fi55
ret_p5cc_gu40 = (1 + p5_gu40).cumprod()  
ret_p20cc_gu40 = (1 + p20_gu40).cumprod()  
ret_p35cc_gu40 = (1 + p35_gu40).cumprod()  
ret_p5fi_gu55 = (1 + p5_gu55).cumprod()  
ret_p20fi_gu55 = (1 + p20_gu55).cumprod()  
ret_p35fi_gu55 = (1 + p35_gu55).cumprod()

# Student cc40 and fi55
ret_p5cc_st40 = (1 + p5_st40).cumprod()  
ret_p20cc_st40 = (1 + p20_st40).cumprod()  
ret_p35cc_st40 = (1 + p35_st40).cumprod()  
ret_p5fi_st55 = (1 + p5_st55).cumprod()  
ret_p20fi_st55 = (1 + p20_st55).cumprod()  
ret_p35fi_st55 = (1 + p35_st55).cumprod()  

# SP500
ret_RmRf = (1 + RmRf).cumprod()
```
